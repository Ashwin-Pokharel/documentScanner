```python
import pandas as pd
import numpy as np
import PIL as pil
import cv2
import tensorflow as tf
import torch as pytorch
import sys
import requests
import tarfile
import json
import numpy as np
from os import path
from PIL import Image
from PIL import ImageFont, ImageDraw
from glob import glob
from matplotlib import pyplot as plt
from pdf2image import convert_from_path
from IPython.display import display
from tensorflow import keras

%matplotlib inline

pil_image = convert_from_path("test_form.pdf")
```

```python
#convert pdf's to images

pil_image = convert_from_path("test_form.pdf")
fig = plt.figure(figsize = (20 , 20))
kernel = np.ones((5,5), np.uint8) 
for i in range(len(pil_image)):
    open_cvImage = np.array(pil_image[i])
    open_cvImage = open_cvImage[: , : , ::-1].copy()
    imgray = cv2.cvtColor(open_cvImage , cv2.COLOR_BGR2GRAY)
    newImg = cv2.bilateralFilter(imgray ,9 ,100 ,100 )
    newImg = cv2.Canny(newImg , 100 , 200)
    newImg = cv2.dilate(newImg , kernel , iterations = 6)
    ret,thresh = cv2.threshold(newImg,127,255,0)
    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    print(len(contours))
    cv2.drawContours(open_cvImage , contours,-1 , (0 , 255 , 0) , 3)
    fig.add_subplot(1 , 2 , i+1)
    plt.imshow(open_cvImage)

plt.show()

```
